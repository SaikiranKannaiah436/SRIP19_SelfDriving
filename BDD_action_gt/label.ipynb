```python
def count(data):
    count_list = []
    for item in data:
        count_dict = {'car':0,'person':0, 'id':item['name'], 'scene':item['attributes']['scene']}
        for label in item['labels']:
            if 'car' in label['category']:
                count_dict['car'] += 1
            elif 'person' in label['category']:
                count_dict['person'] += 1
        count_list.append(count_dict)
    
    return count_list       
```

```python
def threshold(num_p, num_v, data):
    count_list = count(data)
    new_list = []
    id_list = []
    for item in count_list:
        if item['car'] >= num_v and item['person'] >= num_p:
            new_list.append(item)
            id_list.append(item['id'])
    
    return new_list, id_list
```
